```python
import matplotlib.pyplot as plt
from matplotlib import cm
from time import sleep
import numpy as np
%matplotlib inline
```

```python
#rgb
cmap = cm.get_cmap('viridis')
```

```python
Zj = 1
for i in np.linspace(0,0.3,20):
    rgba = cmap(i)
    c = rgba[:-1]
    h,w = 50, 50
    img = np.tile(c, (h,w,1))
    plt.imshow(img)
    plt.show()
    plt.axis('off')
    j += 1
```
